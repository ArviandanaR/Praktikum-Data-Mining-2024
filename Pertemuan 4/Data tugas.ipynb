```python
import pandas as pd
```

```python
df = pd.read_csv('jabodetabek_house_price.csv')
df.head()
```

```python
df.drop(['url', 'title'], axis = 1, inplace = True)
df.head()
```

```python
list(df.columns)
```

```python
df['furnishing'].unique()
```

```python
df.drop(['district', 'city', 'lat', 'long', 'property_type', 'ads_id', 'building_age'], axis = 1, inplace = True)
```

```python
df['facilities'] = df['facilities'].str.lower()
df['facilities'] = df['facilities'].str.replace('jogging track', 'track lari')
df['facilities'] = df['facilities'].str.replace('keamanan 24 jam', 'keamanan')
df['facilities'] = df['facilities'].str.replace('sistem', 'system')
df['facilities'] = df['facilities'].str.replace('onegate', 'one gate system')
df['facilities'] = df['facilities'].str.replace('smarthome', 'smart home')
df['facilities'] = df['facilities'].str.replace('2 unit ac', 'ac 2 unit')
df['facilities'] = df['facilities'].str.replace('1 unit ac', 'ac 1 unit')
df['facilities'] = df['facilities'].str.replace('cctv area', 'cctv')
df['facilities'] = df['facilities'].str.replace('servis', 'service')
df['facilities'] = df['facilities'].str.replace('security24 ', 'security 24')
df['facilities'] = df['facilities'].str.replace('securty', 'security')
df['facilities'] = df['facilities'].str.replace('24h', '24 jam')
df['facilities'] = df['facilities'].str.replace('24jam', '24 jam')
df['facilities'] = df['facilities'].str.replace('swimming pool', 'kolam renang')
df['facilities'] = df['facilities'].str.replace('torn', 'toren')
df['facilities'] = df['facilities'].str.replace('650ltr', '650 liter')
df['facilities'] = df['facilities'].str.replace('torrent', 'toren air')
```
