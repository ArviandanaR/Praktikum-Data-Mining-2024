# Read dataset in JSON

```python
df = pd.read_json('Data pertemuan 5.json')
df.head()
```

# Preprocessing

```python
# Tentukan variabel numerik
X_numeric = df[['Age', 'Fare']]

# Tentukan variabel kategorik
X_categoric = df[['Sex', 'Cabin', 'Embarked', 'Pclass', 'SibSp', 'Parch']]

# Tentukan variabel target
y = df['Survived']
```

```python
# Melihat jumlah missing value

X_numeric.isna().sum()
```

```python
# Melihat jumlah missing value

X_categoric.isna().sum()
```

```python
# Mengisi missing value pada variabel Age dengan nilai rata-rata

mean = X_numeric.Age.mean()
X_numeric.Age = X_numeric.Age.fillna(mean)
```

```python
# Mengisi missing value pada variabel Cabin dengan kategori baru
X_categoric['Cabin'] = X_categoric['Cabin'].fillna('Other')

# Mengisi missing value pada variabel Embarked nilai modus
mode = X_categoric['Embarked'].mode()[0]
X_categoric['Embarked'].fillna(mode, inplace = True)
```

```python
# Import library untuk membuat dummy

from sklearn.preprocessing import OneHotEncoder
```

```python
# Membentuk variabel dummy dengan OneHotEncoder pada variabel kategorik

ohe = OneHotEncoder(sparse_output = False, handle_unknown = 'ignore')

dummy = ohe.fit_transform(X_categoric)

X_encoded = pd.DataFrame(dummy, columns = ohe.get_feature_names_out())
X_encoded
```

```python
# Menyatukan variabel numerik dengan dummy

X = pd.concat([X_numeric, X_encoded], axis = 1)
X
```

```python
# Melihat apakah masih ada missing value

X.isna().any().all()
```

# Modeling

```python
# Import library untuk memisahkan data menjadi data train dan test

from sklearn.model_selection import train_test_split
```

```python
# Pisahkan data menjadi data train dan test

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 26)
```

## K Nearest Neighbor Classifier

<img src="https://raw.githubusercontent.com/ArviandanaR/Praktikum-Data-Mining-2024/main/Pertemuan%206/asset/KNN%20Source.jpg" alt="KNN image" width="750"/>

KNN adalah algoritma machine learning yang bekerja berdasarkan prinsip bahwa objek yang mirip cenderung berada dalam jarak yang dekat satu sama lain. Dengan kata lain, data yang memiliki karakteristik serupa akan cenderung saling bertetangga dalam ruang fitur (feature space).

Algoritma KNN mengasumsikan bahwa objek yang mirip akan berada dalam jarak yang dekat satu sama lain. KNN menggunakan seluruh data yang tersedia dalam pengambilan keputusan. Ketika ada data baru yang perlu diklasifikasikan, algoritma mengukur tingkat kemiripan atau fungsi jarak antara data baru tersebut dengan data yang sudah ada. Data baru kemudian ditempatkan dalam kelas yang paling banyak dimiliki oleh data tetangga terdekatnya.

```python
# Membuat fungsi untuk memodelkan dan mengevaluasi model tersebut

def train_and_evaluate(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = pd.DataFrame(model.predict(X_test), columns = ['y_pred'])

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    result = {'Accuracy': accuracy, 'Precision': precision, 'Recall': recall, 'F1-score': f1}

    print(classification_report(y_test, y_pred))
    return y_pred, result
```

<img src="https://raw.githubusercontent.com/ArviandanaR/Praktikum-Data-Mining-2024/main/Pertemuan%206/asset/Logit%20Source.png" alt="Logit image" width="750"/>

Regresi Logistik merupakan analisis regresi yang digunakan ketika variabel dependennya berupa biner, atau juga bisa disebut sebagai klasifikasi. Regresi logistik menghitung nilai probabilitas terjadinya kejadian pada variabel respons, kemudian mengkategorikannya berdasarkan *threshold* atau batas nilai, umumnya adalah 0.5.  

Module sklearn menyediakan dua jenis regresi logistik yakni LogisticRegression dan LogisticRegressionCV. Meski terdapat dua jenis fungsi, namun sejatinya kedua fungsi tersebut adalah sama.

# All Model

```python
# Membuat fungsi untuk mengevaluasi semua model secara otomatis

def all_model(list_model, X, y, test_size = 0.2, random_state = None):

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = test_size, random_state = random_state)

    # List kosong untuk menyimpan semua hasil pengukuran
    result = []
    for model in list_model:
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        
        # Hitung nilai berdasarkan model yang sudah di train / fit dan hasil prediksinya
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred)
        recall = recall_score(y_test, y_pred)
        f1 = f1_score(y_test, y_pred)

        # Kumpulkan nilai-nilai di atas menjadi satu
        hasil = {
            'Model' : type(model).__name__,
            'Accuracy' : accuracy,
            'F1 Score' : f1,
            'Precision' : precision,
            'Recall' : recall
        }

        # Simpan nilai model ini dalam list result
        result.append(hasil)
    
    # Ubah menjadi dataframe
    result_all = pd.DataFrame(result)

    return result_all
```

```python
# Tuliskan semua model yang ingin digunakan beserta parameter di dalamnya

list_of_model = [
    LogisticRegression(max_iter = 1000),
    LogisticRegressionCV(max_iter = 10000),
    GradientBoostingClassifier(random_state = 26),
    HistGradientBoostingClassifier(random_state = 26),
    CatBoostClassifier(random_state = 26, logging_level = 'Silent'),
    XGBClassifier(random_state = 26),
    BernoulliNB(),
    GaussianNB(),
    MultinomialNB(),
    KNeighborsClassifier()
]
```

```python
# Jalankan fungsi yang telah dibuat

all_listed_model = all_model(list_of_model, X, y, test_size = 0.25)
all_listed_model
```

```python
# Urutkan berdasarkan nilai akurasi

all_listed_model.sort_values('Accuracy', ascending = False)
```

```python
# Buat fungsi untuk melakukan CV terhadap semua fungsi

def all_model_cv(list_model, metric_list, X, y, random_state = None, n_split = 5):
    kfold = KFold(n_splits = n_split, shuffle = True, random_state = random_state)

    result = []
    for model in list_model:

        score = []
        for metric in metric_list:
            metric_score = cross_validate(model, X, y, cv = kfold, scoring = metric)
            score.append(metric_score['test_score'].mean())
        
        result.append(score)
    
    result_all = pd.DataFrame(result, columns = metric_list)

    return result_all
```

```python
# Tentukan metric yang akan digunakan

list_of_metric = [
    'accuracy',
    'f1',
    'recall',
    'precision',
    'roc_auc',
    'neg_log_loss',
    'f1_weighted'
]
```

```python
# Jalankan fungsi untuk menjalankan CV terhadap semua model

all_listed_model_cv = all_model_cv(list_of_model, list_of_metric, X, y, n_split = 5, random_state = 26)
all_listed_model_cv
```

```python
# Tambahkan kolom berisikan nama model dan percantik tampilan dataframe

column_name = [metric.capitalize() for metric in list_of_metric]
model_name = [type(model).__name__ for model in list_of_model]

all_listed_model_cv.columns = column_name
all_listed_model_cv.insert(0, 'Model', value = model_name)
all_listed_model_cv
```
